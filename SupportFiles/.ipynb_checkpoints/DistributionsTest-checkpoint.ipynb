## GENERATE RANDOM LIGHTNING VALUES IN A NORMAL DISTRIBUTION

Import libraries

```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
import random
```

Generate Normal (Gaussian) distribution

```python
x = np.arange(-3, 3, 0.001)
plt.plot(x, norm.pdf(x))
```

Generate random numbers in a normal distribution using the Numpy specific function

"mu" is the desired means
"sigma" is the standard deviation

```python
mu = 10.0
sigma = 4.0
values = np.random.normal(mu,sigma,10000)
plt.hist(values,100)
plt.show()
print(values)
type(values)
```

```python
I_first = 31000
I = np.arange(2000,200000,0.1)
P = 1/(1+((I/I_first)**2.6))
plt.xlim(2000,200000)
plt.plot(I,P)
plt.show()
print(random.choice(P))
```

```python
Z = [10,20,40]
weights = [.2,.6,.2]
random.choices(Z,weights)
```
